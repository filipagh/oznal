```python
x = df['vlhkost vzduchu']
y = df['viditelnost']

z = np.polyfit(x, y, 1 )
ypred = np.polyval(z,x)
residual = np.sum((y-ypred)**2)/len(y)
print(residual)
import matplotlib.pyplot as plt

p = np.poly1d(z)
p30 = np.poly1d(np.polyfit(x, y, 5))
xp = np.linspace(20, 100, 100)
_ = plt.plot(x, y, '.', xp, p(xp), '--', xp, p30(xp), '-')
print(z)
```

```python
import statsmodels.api as sma

linear_regressor = LinearRegression()  # create object for the class
X = df['vlhkost vzduchu'].values.reshape(-1, 1)  # values converts it into a numpy array
Y = df['viditelnost'].values.reshape(-1, 1)
model=linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()

X = sma.add_constant(X)
ss= sma.OLS(Y,X)
sd=ss.fit()
print(sd.summary())
```
